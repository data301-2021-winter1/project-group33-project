```python
import project_functions1
import pandas as pd
import numpy as np
```

```python
fireData = project_functions1.loadAndProcessFireData("../../data/raw")
ONIData = project_functions1.loadAndProcessONIData("../../data/raw/ONI_Data.txt")
dataByMonth = project_functions1.createDataByMonth(fireData, ONIData)

pd.set_option('display.min_rows', 30)
pd.set_option('display.max_rows', 100)
display(dataByMonth)

dataByMonth
```

```python
dataByMonth = dataByMonth.loc[lambda x : (x['Month'] == 'May') | (x['Month'] == 'June') | (x['Month'] == 'July') | (x['Month'] == 'August')]

def anomalyIndex(x1, x2):
    return np.log(x1 / x2)

means = pd.DataFrame(columns = {"Month", "Jurisdiction", "MeanNum", "MeanArea"})
for month in dataByMonth["Month"].unique()[:]:
    for province in dataByMonth["Jurisdiction"].unique():
        means = means.append({
            "Month" : month,
            "Jurisdiction" : province,
            "MeanNum" : dataByMonth[(dataByMonth['Month'] == month) & (dataByMonth['Jurisdiction'] == province)]['Number'].mean(),
            "MeanArea" : dataByMonth[(dataByMonth['Month'] == month) & (dataByMonth['Jurisdiction'] == province)]['Area (hectares)'].mean()  
        }, ignore_index=True)
            
dataBySummerMonth = (
    dataByMonth.merge(
        means,
        how = "left",
        on = ["Month", "Jurisdiction"]
    ).assign(anomalyArea = lambda x: anomalyIndex(x["Area (hectares)"], x["MeanArea"]))
    .loc[lambda x: x['Jurisdiction'] != 'Prince Edward Island']
)

dataBySummerMonth
```

```python
dataBySummerMonth.to_csv("../../data/processed/dataByMonth.csv")
```
